```python
from umift.processing.zarr_dataset import coinft2tcp
import pickle

umift_repo_path = "/local/real/chuerpan/repo/UMI-FT"
gripper_width_data = pickle.load(open(f'{umift_repo_path}/gripper_width_data_batch1.pkl', 'rb'))

ft_data_right = pickle.load(open(f'{umift_repo_path}/ft_data_right_batch1.pkl', 'rb'))
ft_data_left = pickle.load(open(f'{umift_repo_path}/ft_data_left_batch1.pkl', 'rb'))
campose_data = pickle.load(open(f'{umift_repo_path}/campose_data_batch1.pkl', 'rb'))

ft_tcp_left, ft_tcp_right, ft_tcp_conca = coinft2tcp(gripper_width_data, ft_data_left, ft_data_right, campose_data)
```

```python
ft_data_left['ftData'].shape
```

```python
import matplotlib.pyplot as plt

plt.figure()
fig, axes = plt.subplots(4,1)


ax1, ax2, ax3, ax4 = axes 
ax1.plot(ft_data_left['ftData'][..., :3], '.-')
ax1.set_title('ft_data_left')
ax1.legend(["x", "y", "z"])

ax2.plot(ft_data_right['ftData'][..., :3], '.-')
ax2.set_title('ft_data_right')
ax2.legend(["x", "y", "z"])


ax3.plot(ft_data_left['ftData'][..., 3:], '.-')
ax3.set_title('Moment_left')
ax3.legend(["Mx", "My", "Mz"])

ax4.plot(ft_data_right['ftData'][..., 3:], '.-')
ax4.set_title('Moment_right')
ax4.legend(["Mx", "My", "Mz"])

plt.show()


 

```

```python
import matplotlib.pyplot as plt

plt.figure()
fig, axes = plt.subplots(3, 1)


ax1, ax2, ax3 = axes
ax1.plot(ft_tcp_left[..., :3], '.-')
ax1.set_title('ft_tcp_left')
ax1.legend(["x", "y", "z"])

ax2.plot(ft_tcp_right[..., :3], '.-')
ax2.set_title('ft_tcp_right')
ax2.legend(["x", "y", "z"])

ax3.plot(ft_tcp_right[..., :3]+ft_tcp_left[..., :3], '.-')
ax3.set_title('estimated_external_force')
ax3.legend(["x", "y", "z"])

plt.show()

plt.figure()
fig, axes = plt.subplots(3, 1)


ax1, ax2, ax3 = axes
ax1.plot(ft_tcp_left[..., 3:], '.-')
ax1.set_title('T_tcp_left')
ax1.legend(["x", "y", "z"])

ax2.plot(ft_tcp_right[..., 3:], '.-')
ax2.set_title('T_tcp_right')
ax2.legend(["x", "y", "z"])

ax3.plot(ft_tcp_right[..., 3:]+ft_tcp_left[..., 3:], '.-')
ax3.set_title('estimated_external_T')
ax3.legend(["x", "y", "z"])

plt.show()



```
